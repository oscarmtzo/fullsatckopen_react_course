# DOM - Document Object Model - API del navegador

- permite la manipulación programatica del arbol de elementos correspondientes a una página web

- como ejemplo podemos manipular los elementos desde la consola:
    ![devtools DOM document object](./images/00_devtools_DOM_document_object.png)

```python
%%javascript
console.dir(document.documentElement)
```

```python
from IPython.display import display, HTML

# Captura el árbol del DOM directamente usando una propiedad nativa de Chromium
display(HTML("<details><summary><b>Ver DOM Completo (Haz clic para desplegar)</b></summary><pre>" + 
             "&lt;html&gt;\n" + document.documentElement.innerHTML.replace("<", "&lt;") + "\n&lt;/html&gt;" + 
             "</pre></details>"))
```
